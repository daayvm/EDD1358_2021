```
#Array (Arreglo)

class Array:
  def __init__(self, tam):
    self.__datos = [0 for x in range(tam)] #generador, crea un arreglo de 0s

  def get_item(self,index):
    dato = 0
    if index >= 0 and index < len(self.__datos):
      dato = self.__datos[index]
    else:
      print("Error en el indice")
    return dato
  def set_item(self, dato ,index):
    if index >= 0 and index < len(self.__datos):
      self.__datos[index] = dato
    else:
      print("Error en el indice")
  def get_length(self):
    return len(self.__datos)
  def clear (self,dato):
    self.__datos=[dato for x in range(len(self.__datos))]
    return self.__datos
  def getIterator(self):
    return _getIterator(self.__datos)

class _getIterator:
  def __init__( self , arreglo ):
    self.__arreglo = arreglo
    self.__indice_actual = 0

  def __iter__( self ):
    return self

  def __next__( self  ):
    if self.__indice_actual < len(self.__arreglo):
      dato = self.__arreglo[self.__indice_actual]
      self.__indice_actual +=1
      return dato
    else:
      raise StopIteration
  def getIterator( self ):
    return _getIterator( self.__datos )





#Array2D

class Array2D:

    def __init__(self, renglones, columnas):
        self._reng = renglones
        self._col = columnas
        self._array = [[0 for y in range(self._col)] for x in range(self._reng)]

    def clear(self, dato):
        for ren in range(self._reng):
            for col in range(self._col):
                self._array[ren][col] = dato

    def get_tam_reng(self):
        return self._reng

    def get_tam_col(self):
        return self._col

    def set_item(self, reng, col, dato):
        self._array[reng][col] = dato

    def get_item(self, reng, col):
        return self._array[reng][col]

    def to_string(self):
        return self._array
```

```
##Listas Simplemente Ligadas.

class nodo:

    def __init__(self, dato, siguiente=None):  # Argumento con un valor asignado
        self.dato = dato
        self.siguiente = siguiente


class Linkedlist:
    def __init__(self):
        self.__head = None

    def is_empty(self):
        return self.__head == None

    def tail(self):
        curr_node = self.__head
        while curr_node != None and curr_node.siguiente != None:
            curr_node = curr_node.siguiente
        return curr_node

    def append(self, value):
        nuevo = nodo(value)
        if self.__head == None:
            self.__head = nuevo
        else:
            self.tail().siguiente = nuevo

    def transversal(self):
        curr_node = self.__head
        if curr_node == None:
            print("Lista vacia")
        else:
            while curr_node != None:
                print(f" {curr_node.dato} ---->", end="")
                curr_node = curr_node.siguiente
        print("")

    def prepend(self, value):
        nuevo = nodo(value, self.__head)
        self.__head = nuevo

    def remove(self, valor):
        curr_node = self.__head
        while curr_node != None:
            if self.__head.dato == valor:
                self.__head = self.__head.siguiente
                break
            if curr_node.siguiente != None and curr_node.siguiente.dato == valor:
                anterior = curr_node
                anterior.siguiente = curr_node.siguiente.siguiente
                break
            curr_node = curr_node.siguiente

    def pop(self, pos=None):
        curr_node = self.__head
        cont = 0
        while curr_node != None:
            cont += 1
            if pos != None:
                if pos - 1 == cont:
                    anterior = curr_node
                    anterior.siguiente = curr_node.siguiente.siguiente
                    break
                elif pos == 1:
                    self.__head = self.__head.siguiente
                    break
            if pos == None:
                if curr_node.siguiente == None:
                    self.remove(curr_node.dato)
            curr_node = curr_node.siguiente

    def len(self):
        curr_node = self.__head
        cont = 0
        while curr_node != None:
            cont += 1
            curr_node = curr_node.siguiente
        return cont




##Lista Doblemente Ligada

class NodoDoble:
    def __init__(self, value, previo=None, siguiente=None):
        self.data = value
        self.prev = previo
        self.next = siguiente

class DoubleLinkedList:
    def __init__(self):
        self.__head = None
        self.__tail = None
        self.__size = 0

    def get_tail_data(self):
        return self.__tail.data

    def get_size(self):
        return self.__size

    def is_empty(self):
        return self.__size == 0

    def append(self, value):
        if self.is_empty():
            nuevo = NodoDoble(value)
            self.__head = nuevo
            self.__tail = nuevo
        else:
            nuevo = NodoDoble(value, self.__tail)
            self.__tail.next = nuevo
            self.__tail = nuevo
        self.__size += 1

    def transversal(self):
        curr_node = self.__head
        while curr_node != None:
            print("<---", curr_node.data, "---->", end="")
            curr_node = curr_node.next
        print("")

    def reverse_transversal(self):
        curr_node = self.__tail
        while curr_node != None:
            print("<---", curr_node.data, "---->", end="")
            curr_node = curr_node.prev
        print("")

    def find_from_head(self, value):
        curr_node = self.__head
        cont = 0
        while curr_node != None:
            cont += 1
            if curr_node.data == value:
                break
            curr_node = curr_node.next
        return cont

    def find_from_tail(self, value):
        curr_node = self.__tail
        cont = 0
        while curr_node != None:
            cont += 1
            if curr_node.data == value:
                break
            curr_node = curr_node.prev
        return cont

    def remove_from_head(self, value):
        curr_node = self.__head
        while curr_node != None:
            if self.__head.data == value:
                self.__head = self.__head.next
                self.__head.prev = None
                break
            if curr_node.next != None and curr_node.next.data == value:
                anterior = curr_node
                anterior.next = curr_node.next.next
                anterior.next.prev = anterior
                break
            if value == self.__tail.data and curr_node == self.__tail:
                self.__tail = self.__tail.prev
                self.__tail.next = None
            curr_node = curr_node.next

    def remove_from_tail(self, value):
        curr_node = self.__tail

        while curr_node != None:
            if self.__tail.data == value:
                self.__tail = self.__tail.prev
                self.__tail.next = None
            if curr_node.prev != None and curr_node.prev.data == value:
                anterior = curr_node
                anterior.prev = curr_node.prev.prev
                anterior.prev.next = anterior
                break
            if value == self.__head.data and curr_node == self.__head:
                self.__head = self.__head.next
                self.__head.prev = None
            curr_node = curr_node.prev
```

```
##Lista Circular.

class Nodo:
    def __init__(self, value, siguiente=None):
        self.data = value
        self.next = siguiente

class CircularList:
    def __init__(self):
        self.__ref = None
        self.__size = 0

    def is_empty(self, ):
        return self.__size == 0

    def get_size(self):
        return self.__size

    def insert(self, value):
        curr_node = self.__ref
        if curr_node == None:
            nuevo = Nodo(value)
            self.__ref = nuevo
            nuevo.next = nuevo
            self.__size += 1
        elif value > curr_node.data:
            anterior = curr_node.next
            nuevo = Nodo(value)
            curr_node.next = nuevo
            curr_node.next.next = anterior
            self.__ref = curr_node.next
            self.__size += 1
        elif value < curr_node.data:
            if value < curr_node.next.data:
                nuevo = Nodo(value)
                siguiente = curr_node.next
                curr_node.next = nuevo
                curr_node.next.next = siguiente
                self.__size += 1
            else:
                nuevo = Nodo(value)
                curr_node = self.__ref.next
                while curr_node != self.__ref:
                    if curr_node.data < value and curr_node.next.data > value:
                        siguiente = curr_node.next
                        curr_node.next = nuevo
                        curr_node.next.next = siguiente
                        self.__size += 1
                        break
                    curr_node = curr_node.next

    def transversal(self):
        curr_node = self.__ref
        print(f"{curr_node.data}---->",end="")
        curr_node = self.__ref.next
        while curr_node != self.__ref:
            if curr_node.next == self.__ref:
                print(f"{curr_node.data}---->{curr_node.next.data}", end="")
            else:
                print(f"{curr_node.data}---->", end="")
            curr_node = curr_node.next
        print("")

    def search(self,value):
        curr_node = self.__ref.next
        validacion = False
        cont = self.get_size()
        while curr_node != self.__ref or cont > 0:
            if curr_node.data == value :
                validacion = True
            curr_node = curr_node.next
            cont = cont - 1
        return validacion

    def remove(self,value):
        curr_node = self.__ref
        salida = True
        if self.search(value):
            while salida:
                if curr_node.next.data == value:
                    if value == self.__ref.data:
                        curr_node.next = curr_node.next.next
                        self.__ref = curr_node
                        salida = False
                    else:
                        curr_node.next = curr_node.next.next
                        salida = False
                curr_node = curr_node.next
        else:
            print("No se encuentra el valor en la lista")
```

```
##Pila

class Stack:
    def __init__(self):
        self.__data = []
        self.__size = 0

    def pop(self):
        return self.__data.pop()

    def get_size(self):
        return self.__size

    def peak(self):
        if len(self.__data) > 0:
            return self.__data[-1]
        else:
            return None

    def push(self,value):
        self.__data.append(value)
        self.__size += 1

    def to_string(self):
        print("-"*6)
        for dato in self.__data[::-1]:
            print(f"| {dato} |")
            print("-" * 6)
        print("")
```

```
##Cola

class Queue:
    def __init__(self):
        self.__data = list()

    def is_empty(self):
        return len(self.__data) == 0

    def length(self):
        return len(self.__data)

    def enqueue(self, item):
        self.__data.append(item)

    def dequeue(self):
        if not self.is_empty():
            return self.__data.pop(0)
        else:
            return None

    def to_String(self):
        for elem in self.__data:
            print(f"| {elem} ",end="")
        print("|")

##Cola de prioridad no acotada.

class PriorityQueue:
    def __init__(self):
        self.__data = list()

    def is_empty(self):
        return len(self.__data) == 0

    def length(self):
        return len(self.__data)

    def enqueue(self, item,prioridad):
        tupla_nueva = (item, prioridad)
        lista_base = self.__data
        if self.is_empty():
            self.__data.append(tupla_nueva)
        else:
            for pos in range(len(lista_base)):
                if pos != len(self.__data)-1:
                    if prioridad == self.__data[pos][1]:
                        if self.__data[pos][1] == self.__data[pos+1][1]:
                            pass
                        else:
                            self.__data.insert(pos+1,tupla_nueva)
                            break
                    elif prioridad < self.__data[pos][1]:
                        lista_nueva = [tupla_nueva]
                        self.__data = lista_nueva + self.__data
                        break
                else:
                    if prioridad >= self.__data[pos][1]:
                        self.__data.append(tupla_nueva)
                    elif prioridad < self.__data[pos][1]:
                        lista_nueva = [tupla_nueva]
                        self.__data = lista_nueva + self.__data
                        break

    def dequeue(self):
        if not self.is_empty():
            return self.__data.pop(0)
        else:
            return None

    def to_string(self):
        cadena = ""
        for elem in self.__data:
            cadena = cadena + "| " + str(elem[0]) + ", Prioridad: " + str(elem[1])
        cadena = cadena + " |"
        print(cadena)

##Cola de prioridad acotada.

class BoundedPriorityQueue:
    def __init__(self, niveles):
        self.__data = [Queue() for x in range(niveles)]
        self.__size = 0

    def is_empty(self):
        return self.__size == 0

    def queues_are_empty(self):
        FilasVacias = True
        for fila in self.__data:
            if not fila.is_empty():
                FilasVacias = False
        return FilasVacias

    def length(self):
        return self.__size

    def enqueue(self, prioridad, elem):
        if prioridad < len(self.__data) and prioridad >= 0:
            self.__data[prioridad].enqueue(elem)
            self.__size += 1
        else:
            print("Prioridad fuera de rango.")

    def dequeue(self):
        if not self.is_empty():
            nivel = 0
            while self.__data[nivel].is_empty():
                nivel += 1
            return self.__data[nivel].dequeue()

    def to_String(self):
        for nivel in range(len(self.__data)):
            print(f"Nivel {nivel} --> {self.__data[nivel].to_String()}")
```
